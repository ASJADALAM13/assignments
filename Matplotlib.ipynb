```python
#2.
"""Scatter Plot is a plot that helps in visualization of data by adding points based on two variables"""
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plot=plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Data')
plt.show()
```

```python
#5.

"""
Imagine you are a teacher, and you’ve given a test to two different groups of students. After grading, 
you want to see how well each group performed overall, but instead of looking at each individual score,
you want a quick summary that shows how the scores are spread out. This is where a box plot becomes really helpful.

Think of a box plot like a summary of all the test scores:
The middle of the box shows the median score, or the score in the middle of the group.
The edges of the box tell you where most of the scores are (between the 25th and 75th percentile). This helps you see how spread out or close the scores are.
The lines (whiskers) extending from the box show the lowest and highest scores, except for any unusual scores (outliers).
Outliers, or unusual scores, are points far away from the box and are easy to spot.
"""


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2])
```
