```python
#1

""" Five plots that we can plot using the Seaborn library are:-

Bar Plot: Displays the distribution of categorical data.
sns.barplot()
Line Plot: Shows the relationship between two variables with a line.
sns.lineplot()
Scatter Plot: Visualizes the relationship between two continuous variables.
sns.scatterplot()
Box Plot: Displays the distribution of data based on quartiles.
sns.boxplot()
Heatmap: Represents data in a matrix form with colors indicating values.
sns.heatmap()
"""
```

```python
#2.
import matplotlib.pyplot as plt
import seaborn as sns
fmri=sns.load_dataset('fmri')
print(fmri.head())
sns.lineplot(x=fmri.timepoint , y = fmri.signal,hue=fmri.event,style=fmri.region)
```

```python
#3.
titanic=sns.load_dataset('titanic')
print(titanic)
plt.subplot(1,2,1)
sns.boxplot(x=titanic.pclass,y=titanic.age)
plt.subplot(1,2,2)
sns.boxplot(x=titanic.pclass,y=titanic.fare)
```

```python
#4.
diamond=sns.load_dataset('diamonds')
print(diamond)
sns.histplot(x='price',hue='cut',data=diamond)
```

```python
#5
iris=sns.load_dataset('iris')
print(iris.head())
sns.pairplot(hue='species',data=iris)
```

```python
#6.
import pandas as pd
flights=sns.load_dataset('flights')
print(flights.head())
flight_pvt = flights.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')
sns.heatmap(data=flight_pvt,  annot=True,cmap='YlGnBu',fmt='g')
```
